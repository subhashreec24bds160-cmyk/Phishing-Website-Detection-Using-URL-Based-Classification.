```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("projectdata.csv")

plt.figure(figsize=(12, 8))

# Line plot
plt.subplot(2, 2, 1)
plt.plot(df['NoOfLettersInURL'][:500], df['LetterRatioInURL'][:500])
plt.title("Line Plot: NoOfLettersInURL vs LetterRatioInURL")
plt.xlabel("NoOfLettersInURL")
plt.ylabel("LetterRatioInURL")

# Pie Chart
plt.subplot(2, 2, 2)
pie_values = df['LetterRatioInURL'][:5].abs()
plt.pie(pie_values, labels=[f"Sample {i+1}" for i in range(5)])
plt.title("Pie Chart of First 5 Letter Ratios")

# Bar Chart
plt.subplot(2, 2, 3)
plt.bar(df['NoOfLettersInURL'][:10], df['LetterRatioInURL'][:10])
plt.title("Bar Chart (First 10 Samples)")
plt.xlabel("NoOfLettersInURL")
plt.ylabel("LetterRatioInURL")

# Scatter plot
plt.subplot(2, 2, 4)
plt.scatter(df['NoOfLettersInURL'][:500], df['LetterRatioInURL'][:500])
plt.title("Scatter Plot")
```
